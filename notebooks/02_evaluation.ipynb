```python
def load_trained_artifacts():
    """
    Loads saved processor, model and pipeline
    """
    #Load artifacts from memory
    print("🔄 Loading components: preprocessor and model...")
    preprocessor = joblib.load(os.environ["MODEL_NAME_PREPROCESSOR"])
    model = joblib.load(os.environ["MODEL_NAME_XGB"])

    print("✅ Artifacts successfully loaded!")
    return preprocessor, model
```

```python
preprocessor, model = load_trained_artifacts()
```
