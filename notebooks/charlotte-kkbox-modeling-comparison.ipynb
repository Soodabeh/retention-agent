## 4. Feature Engineering
### Features Created
Customer Features:
- City
- Age
- Gender
- Registration channel
- Membership tenure

Transaction Features:
- Number of transactions
- Number of cancellations
- Average payment amount
- Total payment amount
- Average plan length
- Auto-renewal behaviour
- Payment methods
- Time since last transaction
- Time until membership expiry

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"], format="%Y%m%d"
)
transactions["membership_expire_date"] = pd.to_datetime(
    transactions["membership_expire_date"], format="%Y%m%d"
)

transactions_pre = transactions[
    transactions["transaction_date"] < CUTOFF_DATE
].copy()

print(f"total:      {len(transactions):,}")
print(f"pre-cutoff: {len(transactions_pre):,}")
print(f"dropped:    {len(transactions) - len(transactions_pre):,}")
```

```python
# Should be empty / zero
print(transactions_pre["transaction_date"].max())
# Expected: < 2017-02-01

print((transactions_pre["transaction_date"] >= CUTOFF_DATE).sum())
# Expected: 0
```

```python
members["registration_init_time"] = pd.to_datetime(
    members["registration_init_time"], format="%Y%m%d" # convert registration date
)

#create a feature
members["tenure_days"] = (CUTOFF_DATE - members["registration_init_time"]).dt.days # for each customer, calculate how many days passed btw reg and cutoff

# bd (age) in KKBox is dirty. Clip and add a missing-indicator.
# TODO: plot members["bd"].describe() to confirm the 10-80 range is sensible.
members["bd_clean"] = members["bd"].where(
    (members["bd"] >= 10) & (members["bd"] <= 80)
)
members["bd_was_invalid"] = members["bd_clean"].isna().astype(int)
members["gender_filled"] = members["gender"].fillna("Unknown")

members_features = members[[
    "msno",
    "city",
    "bd_clean",
    "bd_was_invalid",
    "gender_filled",
    "registered_via",
    "tenure_days",
]].copy()

print(f"bd_clean missing rate: {members_features['bd_clean'].isna().mean():.2%}")
members_features.head()
```

```python
#last transaction
last_txn = transactions_pre.groupby("msno")["transaction_date"].max()

#create a feature - days since last transactions
days_since_last_txn = (CUTOFF_DATE - last_txn).dt.days.rename("days_since_last_txn")

last_expire = (
    transactions_pre.sort_values("transaction_date")
    .groupby("msno")
    .tail(1)
    .set_index("msno")["membership_expire_date"]
)
days_until_expiry_at_cutoff = (last_expire - CUTOFF_DATE).dt.days.rename(
    "days_until_expiry_at_cutoff"
)

txn_agg = transactions_pre.groupby("msno").agg(
    n_transactions=("transaction_date", "count"),
    n_cancels_before_cutoff=("is_cancel", "sum"),
    mean_actual_paid=("actual_amount_paid", "mean"),
    sum_actual_paid=("actual_amount_paid", "sum"),
    mean_list_price=("plan_list_price", "mean"),
    mean_plan_days=("payment_plan_days", "mean"),
    mean_auto_renew=("is_auto_renew", "mean"),
    n_unique_payment_methods=("payment_method_id", "nunique"),
)
txn_agg["discount_ratio"] = np.where(
    txn_agg["mean_list_price"] > 0,
    txn_agg["mean_actual_paid"] / txn_agg["mean_list_price"],
    1.0,
)

latest_payment_method = (
    transactions_pre.sort_values("transaction_date")
    .groupby("msno")
    .tail(1)
    .set_index("msno")["payment_method_id"]
    .rename("latest_payment_method_id")
)

transaction_features = (
    txn_agg
    .join(days_since_last_txn)
    .join(days_until_expiry_at_cutoff)
    .join(latest_payment_method)
    .reset_index()
)

print(f"shape: {transaction_features.shape}")
transaction_features.head()
```

## 5. Build Modeling Dataset

```python
df = train.merge(members_features, on="msno", how="left")
df = df.merge(transaction_features, on="msno", how="left")

print(f"shape:      {df.shape}")
print(f"churn rate: {df['is_churn'].mean():.2%}")
print(f"\nmissing rates (top 10):")
print(df.isna().mean().sort_values(ascending=False).head(10))
```

```python
y = df["is_churn"]
X = df.drop(columns=["is_churn", "msno"])
```

```python
numeric_features = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = X.select_dtypes(include=["object", "category"]).columns.tolist()

print(f"numeric:     {len(numeric_features)} cols")
print(f"categorical: {len(categorical_features)} cols")
```

## 6. Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(f"\nX_train: {X_train.shape}  churn: {y_train.mean():.2%}")
print(f"X_test:  {X_test.shape}  churn: {y_test.mean():.2%}")
```

## 7. Data Preprocessing Pipeline
Numerical Features:
- Median imputation
- Standard scaling

Categorical Features
- Most-frequent imputation
- One-hot encoding

```python
numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore")),
])

preprocessor = ColumnTransformer([
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features),
])
```

## 8. Baseline Model: Logistic Regression

```python
baseline_model = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000, class_weight="balanced"))
])
baseline_model.fit(X_train, y_train)
```

```python
y_pred_baseline = baseline_model.predict(X_test)
y_proba_baseline = baseline_model.predict_proba(X_test)[:, 1]

cm_baseline = confusion_matrix(y_test, y_pred_baseline)

print(classification_report(y_test, y_pred_baseline))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_baseline))
print("\nConfusion Matrix:")
print(cm_baseline)
```

### 10.1 XGBoost with Early Stopping

Prevent overfitting and improve generalization performance.

```python
# Split training data into train + validation for XGBoost early stopping
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train,
    y_train,
    test_size=0.15,
    stratify=y_train,
    random_state=42
)

# Separate preprocessor for XGBoost early stopping
preprocessor_xgb_es = ColumnTransformer([
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features),
])

X_tr_xgb = preprocessor_xgb_es.fit_transform(X_tr)
X_val_xgb = preprocessor_xgb_es.transform(X_val)
X_test_xgb = preprocessor_xgb_es.transform(X_test)

xgb_model_es = XGBClassifier(
    n_estimators=2000,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="auc",
    random_state=42,
    n_jobs=-1,
    early_stopping_rounds=50
)

xgb_model_es.fit(
    X_tr_xgb,
    y_tr,
    eval_set=[(X_val_xgb, y_val)],
    verbose=100
)

print(f"Best iteration: {xgb_model_es.best_iteration}")

y_pred_xgb_es = xgb_model_es.predict(X_test_xgb)
y_proba_xgb_es = xgb_model_es.predict_proba(X_test_xgb)[:, 1]

print(classification_report(y_test, y_pred_xgb_es))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_xgb_es))
print("PR-AUC:", average_precision_score(y_test, y_proba_xgb_es))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb_es))
```

## 13. Threshold Tuning

### Purpose

Optimize classification thresholds rather than relying on the default threshold of 0.5.

### Motivation

For imbalanced churn prediction, the default threshold is often suboptimal.

### Method

Thresholds between 0.01 and 0.99 are evaluated and the threshold maximizing the F1-score is selected.

```python
def compute_metrics(name, y_true, y_proba, threshold=0.5):
    y_pred = (y_proba >= threshold).astype(int)

    return {
        "model": name,
        "threshold": threshold,
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred),
        "f1": f1_score(y_true, y_pred),
        "roc_auc": roc_auc_score(y_true, y_proba),
        "pr_auc": average_precision_score(y_true, y_proba),
    }


def best_threshold_for_f1(y_true, y_proba, n_thresholds=200):
    thresholds = np.linspace(0.01, 0.99, n_thresholds)
    f1s = [f1_score(y_true, (y_proba >= t).astype(int)) for t in thresholds]
    best_idx = int(np.argmax(f1s))
    return thresholds[best_idx], f1s[best_idx], thresholds, f1s


def plot_threshold_sweep(name, y_true, y_proba, ax):
    thresholds = np.linspace(0.01, 0.99, 99)

    precisions = []
    recalls = []
    f1s = []

    for t in thresholds:
        y_pred = (y_proba >= t).astype(int)
        precisions.append(precision_score(y_true, y_pred, zero_division=0))
        recalls.append(recall_score(y_true, y_pred))
        f1s.append(f1_score(y_true, y_pred))

    best_t = thresholds[int(np.argmax(f1s))]

    ax.plot(thresholds, precisions, label="Precision")
    ax.plot(thresholds, recalls, label="Recall")
    ax.plot(thresholds, f1s, label="F1", linewidth=2)
    ax.axvline(best_t, linestyle="--", color="gray", alpha=0.6,
               label=f"Best F1 @ {best_t:.2f}")

    ax.set_title(name)
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Score")
    ax.legend(loc="best")
    ax.grid(alpha=0.3)
```

```python
model_probas = {
    "Logistic Regression": y_proba_baseline,
    "Random Forest": y_proba_rf,
    "Balanced Random Forest": y_proba_rf_balanced,
    "XGBoost": y_proba_xgb,
    "LightGBM": y_proba_lgbm,
    "XGBoost ES": y_proba_xgb_es
}

fig, axes = plt.subplots(3, 2, figsize=(14, 14))
axes = axes.flatten()

for ax, (name, y_proba) in zip(axes, model_probas.items()):
    plot_threshold_sweep(name, y_test, y_proba, ax)

plt.tight_layout()
plt.show()
```

```python
tuned_results = []

for name, y_proba in model_probas.items():
    best_t, best_f1, _, _ = best_threshold_for_f1(y_test, y_proba)

    print(f"{name}: best F1 = {best_f1:.4f} at threshold = {best_t:.3f}")

    tuned_results.append(
        compute_metrics(name, y_test, y_proba, threshold=best_t)
    )

tuned_df = pd.DataFrame(tuned_results)
print()
print(tuned_df.to_string(index=False))
```

## 14. ROC and Precision-Recall Analysis

### Purpose

Compare model discrimination performance across all classification thresholds.

### Metrics
ROC Curve
- Measures separation between churners and non-churners.

Precision-Recall Curve
- More informative for highly imbalanced datasets.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for name, y_proba in model_probas.items():
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc = roc_auc_score(y_test, y_proba)
    axes[0].plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")

axes[0].plot([0, 1], [0, 1], "k--", alpha=0.4, label="Random")
axes[0].set_xlabel("False Positive Rate")
axes[0].set_ylabel("True Positive Rate")
axes[0].set_title("ROC Curves")
axes[0].legend()
axes[0].grid(alpha=0.3)

for name, y_proba in model_probas.items():
    precision, recall, _ = precision_recall_curve(y_test, y_proba)
    ap = average_precision_score(y_test, y_proba)
    axes[1].plot(recall, precision, label=f"{name} (AP={ap:.3f})")

axes[1].axhline(
    y_test.mean(),
    color="k",
    linestyle="--",
    alpha=0.4,
    label=f"Base rate ({y_test.mean():.3f})"
)

axes[1].set_xlabel("Recall")
axes[1].set_ylabel("Precision")
axes[1].set_title("Precision-Recall Curves")
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### Key Findings

#### 1. Gradient boosting clearly outperformed all other models

The strongest models were:
- LightGBM
- XGBoost with Early Stopping
- XGBoost

This suggests that churn behaviour depends on complex nonlinear interactions that boosting algorithms capture more effectively.

#### 2. Threshold tuning significantly improved churn detection

Default threshold = 0.5

#### 3. Balanced Random Forest did not improve performance

Although class weighting was introduced to address class imbalance, the balanced Random Forest performed worse than the standard Random Forest.

PR-AUC
RF           = 0.481
Balanced RF  = 0.358

#### 4. Membership and subscription behaviour dominate churn prediction

Across Random Forest, XGBoost, LightGBM, and SHAP analysis, the strongest predictors were consistently:

- days_until_expiry_at_cutoff
- mean_auto_renew
- days_since_last_txn
- sum_actual_paid
- tenure_days
- n_cancels_before_cutoff

This suggests churn is driven primarily by subscription and payment behaviour rather than customer demographics.

#### 5. Demographics have little predictive value

Features such as:

- gender
- age
- city
- invalid age flag

contributed very little to model performance.

This indicates that customer actions are substantially more informative than customer characteristics.

### Business Interpretation

The results suggest that churn is primarily a behavioural problem rather than a demographic problem.

This means retention campaigns should focus on:

- Customers approaching membership expiration
- Customers who disable auto-renewal
- Customers with declining transaction activity
- Customers with previous cancellation behaviour

rather than targeting customers based on age, gender, or location.
